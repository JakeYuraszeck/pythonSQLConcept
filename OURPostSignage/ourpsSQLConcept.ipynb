```python
class ExcelData:
    def __init__(self,fileLocation: str,fileName,fromVisnata: bool) -> None:
        if isinstance(fileName,str):
            self.fullpath = fileLocation +'\\'+fileName

            if fromVisnata :
                self.df= pd.read_csv(self.fullpath,delimiter='\t',encoding_errors='ignore',low_memory=False,keep_default_na=False)
            else:
                self.df = pd.read_excel(self.fullpath,encoding_errors='ignore',low_memory=False,keep_default_na=False)
        elif isinstance(fileName, list):
            self.fullpaths = {}
            dfs = []
            for name in fileName:
                self.fullpaths[name] = fileLocation + '\\' + name
                if fromVisnata:
                    dfs.append(pd.read_csv(self.fullpaths[name],delimiter='\t',encoding_errors='ignore',low_memory=False,keep_default_na=False))
                else:
                    dfs.append(pd.read_excel(self.fullpaths[name],encoding_errors='ignore',low_memory=False,keep_default_na=False))
            self.df = pd.concat(dfs)

        
    def get_df(self):
        return self.df
    
    def get_fullpaths(self):
        return self.fullpaths
    
    def get_rowDict(self,rowNum):
        return self.df.iloc[rowNum].to_dict()
```
